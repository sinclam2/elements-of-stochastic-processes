- How large must $\rho_{UV}$ be before your entire investment would be in $U$?

_____

- First, we'll loop through different scenarios and plot them

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = 10, 10
```

```python
step = 0.01
x = np.arange(0, 1000+step, step)
corr = np.arange(0,1+step, step)

list_results = []

for i, rho in enumerate(corr):
    covariance = rho*np.sqrt(0.15*0.2)
    variances = 0.15*x**2 + 2*x*(1000-x)*covariance + 0.2*(1000-x)**2
    min_x = x[np.argmin(variances)]
    list_results.append(min_x)
```

```python
plt.plot(corr, list_results)
```

```python
corr[np.argmax(list_results)]
```

- As we can see, any correlation value greater than or equal to 0.87 causes us to allocate everything to $U$
    - We'll derive the exact value

$$
Var(P) = 0.15x^{2} + 2\sqrt{0.03}x(1000-x)\rho_{UV} + 0.2(1000-x)^{2} 
$$

- Taking the derivative:

$$
\frac{d}{dx} Var(P) = -4\sqrt{0.03}\rho_{UV}x + 2000\sqrt{0.03}\rho_{UV} +0.7x - 400
$$

- Setting $x$ to 1000 and setting the derivative to zero

$$
0 = -4\sqrt{0.03}\rho_{UV}(1000) + 2000\sqrt{0.03}\rho_{UV} +0.7(1000) - 400
$$

$$
0 = \left(700-400 \right) + 1000\sqrt{0.03}\rho_{UV}(2-4)
$$

$$
2000\sqrt{0.03}\rho_{UV} = 300 \implies \rho_{UV} \approx 0.866
$$
