- Suppose two stocks each cost 1 dollar per share today
- Let the random variables $U$ and $V$ denote their values per share one year from today
- Suppose I know that $Var(U) = 0.15$, $Var(V) = 0.2$, and $U$ and $V$ are uncorrelated
- If I have 1000 dollars to invest today, then we define $x$ such that I invest $x$ dollars into stock $U$ and $1000-x$ into stock $V$
    - Therefore, my profit at the end of the year is equal to: $P = xU + (1000-x)V$
    
- *How much should I invest into each stock to minimize the variance of $P$?*

_____

$$
Var(P) = Var \left ( xU + (1000-x)V \right ) = Cov(xU + (1000-x)V, xU + (1000-x)V) \\
= Cov(xU, xU) + 2Cov(xU, (1000-x)V) + Cov((1000-x)V,(1000-x)V)
$$

$$
= x^{2}Var(U) + 2x(1000-x)Cov(U,V) + (1000-x)Var(V) = 0.15x^{2} + 0.2(1000-x)^{2}
$$

$$
= 0.35x^{2} - 400x + 200000
$$

- Now, taking the derivative and setting to 0

$$
0.7x - 400 = 0 \implies x = \frac{400}{0.7} = 571.43
$$

- So, the variance is minimized when we invest 571.43 into $U$ and 428.57 into $V$

____

- Let's test the variances for different combinations of $U$ and $V$ to confirm our result
